```python
import pandas as pd
import matplotlib.pyplot as plt

from statsmodels.tsa.statespace.sarimax import SARIMAX

import numpy as np
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import root_mean_squared_error, mean_absolute_percentage_error

import warnings
warnings.filterwarnings('ignore')  # Suppress all warnings
```

```python
df = pd.read_csv('EUR_GBP_dataset.csv')
```

```python

time = 'TIME_PERIOD'

target = ['OBS_VALUE_diff']
        # 'OBS_VALUE'


exog_cols = [ 'MONTH', 'WEEK', 'lag_1d', 'lag_2d',
        'rolling_mean_1d', 'rolling_mean_2d',
       'rolling_mean_3d', 'rolling_mean_9d', 'rolling_mean_20d',
       'rolling_mean_60d', 'rolling_std_2d', 'rolling_std_3d',
       'rolling_std_9d', 'rolling_std_20d', 'rolling_std_60d', 'lag_3d',
       'lag_12d']


exog_train = df[exog_cols][:-10] 
exog_test =  df[exog_cols][-10:]
```

```python
train = df[target][:-10] 
test =  df[target][-10:]

mod = SARIMAX(endog = train, 
            order = (1,1,1), 
            seasonal_order=(1,1,1,52),
            exog = exog_train, 
            enforce_stationarity=False,
            enforce_invertibility=False)

print('1')

res  = mod.fit(disp=False)
print('2')
fcst = res.forecast(steps = 10)

print('3')
plt.plot(list(test))
plt.plot(list(fcst))

plt.legend(['Actual data', 'Forecast'])
plt.ylabe('Sales')
plt.xlabel('Test Data Time Step')
```
