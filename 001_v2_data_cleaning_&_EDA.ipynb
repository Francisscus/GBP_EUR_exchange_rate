```python
df_EUR_GBP_estimates = pd.read_csv('EUR_GBP_estimates.csv')
```

```python
df_EUR_GBP_estimates
```

```python
df_EUR_GBP_estimates.info()
```

```python
df_EUR_GBP_estimates['TIME_PERIOD'] = pd.to_datetime(df_EUR_GBP_estimates['TIME_PERIOD'], errors ='coerce')
```

```python
df_EUR_GBP_estimates['YEAR'] = df_EUR_GBP_estimates['TIME_PERIOD'].dt.year
df_EUR_GBP_estimates['MONTH'] = df_EUR_GBP_estimates['TIME_PERIOD'].dt.month
```

```python
df_EUR_GBP_estimates['WEEK'] = df_EUR_GBP_estimates['TIME_PERIOD'].dt.isocalendar().week
```

```python
df_EUR_GBP_estimates
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,5))
plt.plot(  df_EUR_GBP_estimates['TIME_PERIOD'], df_EUR_GBP_estimates['hourly_rate_40_pounds']);
```

```python
plt.figure(figsize=(15,5))
plt.title('Hourly rate 40 GBP in Euros after COVID')
plt.ylabel('Hourly rate 40 GBP in Euros')
plt.xlabel('Time')
plt.plot( df_EUR_GBP_estimates[df_EUR_GBP_estimates['YEAR']>=2022]['TIME_PERIOD'], 
         df_EUR_GBP_estimates[df_EUR_GBP_estimates['YEAR']>=2022]['hourly_rate_40_pounds']);
```

```python
df_EUR_GBP_estimates[df_EUR_GBP_estimates['YEAR']>2023]['hourly_rate_40_pounds'].mean()
```
